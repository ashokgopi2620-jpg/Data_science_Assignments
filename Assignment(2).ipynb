# Estimation And Confidence Intervals

```python
# importing the libraries
import numpy as np 
import pandas as pd
import scipy.stats as st
# sample data
x=[1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
df=pd.DataFrame(x) # convert the given list of sample to Data Frame
df.columns=['sample data']
df
```

```python
df.describe() # To describe the data frame
```

```python
x_bar=df.mean() # mean of the sample data 
x_bar
```

```python
s=df.std() # sample standard deviation
s
```

```python
n=df.count() # length of the sample
n
```

```python
# t-Distribution for sample standard deviation
st.t.interval(0.99,n-1,loc=x_bar,scale=s/np.sqrt(n)) #scale is standard error
```

while using the sample standard deviation ,We are 99% confident that the true average durability of print-heads lies between 1.09019734 million and 1.38713599 million characters before failure.

```python
t=st.t.ppf(1-(0.01)/2,n-1)
t
```

```python
me=t*s/np.sqrt(n) # Margin of error which is error from the sample mean 
me
```

```python
x_bar-me,x_bar+me #manual way 
```

```python
# z-Distribution
sigma=0.2 # population satandard deviation 
st.norm.interval(0.99,loc=x_bar,scale=sigma/np.sqrt(n)) # z distribution
```

while using population standard deviation ,We are 99% confident that the true average durability of print-heads lies between 1.10565141 million and 1.37168192 million characters before failure .
